## 画像処理とは
本章では画像処理について学びます。早速ですが画像処理とは何でしょうか？  
画像処理とは 
**「デジタル画像から何らかの情報を取り出す。もしくは取得しやすくする」**
ために行う処理のことを指します。
当たり前の話ですが、いきなり画像処理を導入しようとしても、そもそものデジタル画像がなかったり、ユースケースによっては定期的に画像が収取できる環境がなければソリューションを実務活用することはできません。

### 画像の数値化
画像処理を学ぶにあたり、画像がどのような形で扱われるのか、どのようにしたら画像から情報を取り出すことができるのか学びましょう。
結論から申し上げますと画像は数値化を行い、数値から様々な情報を得て画像処理のユースケースのような形で扱います。  
では画像の数値化とはどういうことでしょうか、普段我々の目に見えている画像は実は数字の集まりで成り立っています。
![image](../src/img/01/2.PNG)

画像の大きさはそれぞれ高さと横の大きさを表す、ピクセルであらわされます。
またそれぞれのマス目には輝度と呼ばれる、文字通りの明るさを表す数字が0-255 の範囲で入ります。

カラー画像を表す際は、赤、緑、青のチャネルと呼ばれるフィルタにそれぞれの色の輝度が入り、光の三原色によって我々が普段目にする様々な色が表現されています。これが画像の数値化です。この数値情報を処理することによって新たな情報を得たり、数値の特徴を学習し画像AIなどが生成されます。


### 畳み込み
画像×DL（ディープラーニング）の代表的な処理である、畳み込みについて解説します。
この畳み込みをDLに実装したものをCNN（Convolutional Neural Network）と呼びます。

![image3](../src/img/01/3.PNG)
この畳み込みという処理は画像に対し、フィルタと呼ばれるものを掛け合わせて新しい特徴マップを生成する処理のことをさします。
![image4](../src/img/01/4.PNG)
この特徴マップを作ることによって画像の中の輝度の変化点や差がある部分、つまりは物体の輪郭などをとらえやすくすることができるようになります。  


この時、フィルタの値によって特徴マップは大きく変わります。ではどのような値がふさわしいのでしょうか？
![image3](../src/img/01/5.PNG)
この適切なフィルタの値を沢山の画像を読み込ませて計算させて、導き出すことこそが CNN です。  
フィルタの値を未知の値の W とおいて一連の数式を考えると、
y = w1×x1+w2×x2‥とおくことができ、皆さんが高校や大学で学ばれた重回帰分析や単回帰分析などと近い数式を立てることができます。  


本日は入門ということでプーリングやパディング、逆伝播などの難しい説明を省いておりますが根本的な考えとしてはどのように画像を一般的なNNに扱える形にするのかを畳み込みという処理によって実現したものとなっております。


また、画像処理において画像の特徴は輝度から得ていることがわかりました。つまり、野外で撮影した画像などを扱う際には同じ画像でも朝や夜で輝度が異なるため、結果が異なってしまう可能性があることを注意しましょう。
