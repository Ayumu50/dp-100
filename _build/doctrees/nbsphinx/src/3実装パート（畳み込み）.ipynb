```python
# prompt: 画像処理でCNNの畳み込みを行い、畳み込み前と後の画像を比較したい。フィルタの値は個人で定義したい。

import numpy as np
import cv2
import matplotlib.pyplot as plt

# 画像を読み込み
image = cv2.imread("/content/trainocate.png")
img_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)
# フィルタを作成
kernel = np.array([
    [-1, 0, 1],
    [-1, 0, 1],
    [-1, 0, 1]
])
# 数値を変更してみる
#kernel = np.array([
#     [-1, -1, -1],
#     [0, 0, 0],
#     [1, 1, 1]
# ])

# 畳み込みを行う
convolved_image = cv2.filter2D(img_gray, -1, kernel)

# 畳み込み前の画像の表示
plt.imshow(img_gray,cmap='gray')


```

```python
#畳み込み後の画像
plt.imshow(convolved_image,cmap='gray')
```
