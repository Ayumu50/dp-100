# Mod5 ストレージ

## S3
- データレイクとしての活用
ビックデータを扱った分析基盤システムを構築する際のデータレイク（データの一時集計場所としても利用可能）
- [デモ S3静的Webページの公開](https://docs.aws.amazon.com/ja_jp/AmazonS3/latest/userguide/WebsiteHosting.html)  

- すべてのストレージクラスを横断したライフサイクルポリシーの設定が可能  
- バケットポリシーには20kbの制限がある。複数のオブジェクトにアクセス制御をするときにはバケットACLを使用  
- バケットポリシーのConditionにaws:kms を記述することで、アップロード時の暗号化を強制することができる。
- オブジェクトロックで削除・修正をブロックできる。バージョニングと併用する必要がある。
- バッチオペレーションを使用してオブジェクトのコピーはLamda関数のコールなど大量のデータに対する処理を周期的に設定することができる。  
【ポリシーの使い分け】  
**IAM ポリシーはアカウント単位、バケットポリシーはバケット単位、ACLはオブジェクト単位でアクセス制御をすることが多い**

## 暗号化の補足
### データの暗号化とは
暗号化→暗号化キーと暗号アルゴリズムによって解読不可能な暗号文に変換すること。暗号文は複合キーでもとに戻す。暗号化、復号化の双方向の変換ができるので可逆性ともいう。同じ鍵のときは共通鍵暗号方式、別の時を公開暗号方式がある

- ハッシュ化:不可逆な変換、電子署名などで利用されている

KMS は暗号化キーの作成と制御を簡単にする。共通鍵、公開鍵どちらも使用することができる。また、FIPS という米国企画に対応した暗号化がなされている。
